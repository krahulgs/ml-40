```python
import pandas as pd
bike_data=pd.read_csv('C:\\ISHA_STUDY\\bike_sharing_daily.csv')
bike_data
```

```python
bike_data['day']=pd.DatetimeIndex(bike_data['dteday']).day
bike_data
```

```python
data=['temp','atemp','hum', 'windspeed', 'casual']
bike_data[data].describe()
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

var=bike_data['casual']

min_val=var.min()
max_val=var.max()
mean_val=var.mean()
med_val=var.median()
mod_val=var.mode()[0]

#print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                        #mean_val,
                                                                                        #med_val,
                                                                                        #mod_val,
                                                                                        #max_val))

fig=plt.figure(figsize=(9,12))

#adding color to histogram
a=fig.add_subplot(2,1,1)
var.plot.hist(color='lightblue', bins=100)

#plt.hist(var)

plt.axvline(x=min_val,color='red',linestyle='dashed',linewidth=2)
plt.axvline(x=max_val,color='green',linestyle='dashed',linewidth=2)
plt.axvline(x=mean_val,color='red',linestyle='dashed',linewidth=2)
plt.axvline(x=med_val,color='green',linestyle='dashed',linewidth=2)
plt.axvline(x=mod_val,color='white',linestyle='dashed',linewidth=2)

#plt.title('data distribution')
plt.xlabel('value')
plt.ylabel('frequncy')


#boxplot diagram
a=fig.add_subplot(2,1,2)
var.plot(kind='box', vert=False)

fig.show()
```

```python
for col in data:
    fig=plt.figure(figsize=(9,12))

    #adding color to histogram
    a=fig.add_subplot(2,1,1)
    feature=bike_data[col]
    ax=fig.gca()
    feature.plot.hist(color='lightblue', bins=100, ax=ax)
   
    ax.axvline(feature.mean(),color='red',linestyle='dashed',linewidth=2)
    ax.axvline(feature.median(),color='green',linestyle='dashed',linewidth=2)
    
    plt.xlabel(col)
    plt.ylabel('frequncy')
    #a=fig.add_subplot(2,1,2)
    #feature.plot(kind='box', vert=False)
    
plt.show()
    
```
